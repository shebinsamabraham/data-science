### Graduate Program - Master Track in Data Science

The Masters Track in Data Science instills students the skills essential
to knowledge discovery efforts to identify standard, novel and truly
differentiated solutions and decision making, including skills in
managing, querying, analyzing, visualizing, and extracting meaning from
extremely large data sets.

##### DATA SCIENCE TRACK CURRICULUM

this table.
